```python


myList = [1, 2, 3, 4, 5, 6, 7, 8, 9]

mySubList = [2, 6, 5]


pos = -1

def find(item, myList):
  global pos
  j=pos
  for i in range(j+1, len(myList)):
    if item == myList[i]:
      pos = i
      return i
  return -1


def contains(myList, mySubList):
  global pos

  if len(mySubList) > len(myList):
    print("it's Gone")

  for item in mySubList:

    idx = find(item, myList)
    
    if idx == -1:
      print("it's Gone")
      break
  else:
    print("it's a Match")

print(myList)
print(mySubList)
  
contains(myList, mySubList)
```
